```python
import pandas as pd
from pathlib import Path
import datetime
import numpy as np
import os
from zoneinfo import ZoneInfo
```

```python
schedule_data = pd.read_csv("../data/results/schedules.csv")
schedule_data.head()
```

```python
def generate_status_history(schedule_data: pd.DataFrame) -> pd.DataFrame:
    result = {}
    result["status_id"] = []
    result["schedule_id"] = []
    result["start_time"] = []
    result["end_time"] = []
    for _, row in schedule_data.iterrows():
        result["status_id"].append(row["status"])
        result["schedule_id"].append(row["id"])
        result["start_time"].append(row["scheduled_arrival"])
        result["end_time"].append(row["scheduled_departure"])
    return pd.DataFrame(result)
```

```python
status_history = generate_status_history(schedule_data)
status_history
```
